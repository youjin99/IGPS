```python
import pandas as pd
```

```python
df = pd.read_csv('동별_그리드.csv')
```

```python
df
```

```python
df = df[['ADM_DR_NM','fid','rank']]
```

```python
list = df['ADM_DR_NM'].unique()
```

```python
df_sort_group_top12 = df.sort_values(by='rank', ascending = True).groupby('ADM_DR_NM').head(12)
```

```python
df_sort = df_sort_group_top12.sort_values(by=['ADM_DR_NM','rank'],ascending=True)
```

```python
df_sort.to_csv('동별우선순위_lr.csv',encoding = 'cp949')
```
